```python
s=input()
k=3
alph="abcdefghijklmnopqrstuvwxyz"
k=k%len(alph)
s_low=alph[k:]+alph[:k]
s_up=s_low.upper()
table=str.maketrans(alph+alph.upper(),s_low+s_up)
print(s.translate(table))
```
