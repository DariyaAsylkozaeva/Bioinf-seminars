```python
import random
secret=random.randint(1,100)
attempts=10
print('Я загадал число от 1 до 100. У тебя 10 попыток: ')
for attempt in range(1, attempts+1):
    guess=int(input("Попытка {attempt}: "))
    if guess==secret:
        print("Ты выиграл")
        break
    elif guess<secret:
        print("Число больше")  
    else:
        print("Число меньше")    
print("Ты проиграл")      
```

```python
english_scores = {'A': 1, 'E': 1, 'I': 1, 'O': 1, 'U': 1, 'L': 1, 'N': 1, 'S': 1, 'T': 1, 'R': 1,
        'D': 2, 'G': 2,
        'B': 3, 'C': 3, 'M': 3, 'P': 3,
        'F': 4, 'H': 4, 'V': 4, 'W': 4, 'Y': 4,
        'K': 5,
        'J': 8, 'X': 8,
        'Q': 10, 'Z': 10}
russian_scores = {'А': 1, 'В': 1, 'Е': 1, 'И': 1, 'Н': 1, 'О': 1, 'Р': 1, 'С': 1, 'Т': 1,
        'Д': 2, 'К': 2, 'Л': 2, 'М': 2, 'П': 2, 'У': 2,
        'Б': 3, 'Г': 3, 'Ё': 3, 'Ь': 3, 'Я': 3,
        'Й': 4, 'Ы': 4,
        'Ж': 5, 'З': 5, 'Х': 5, 'Ц': 5, 'Ч': 5,
        'Ш': 8, 'Э': 8, 'Ю': 8,
        'Ф': 10, 'Щ': 10, 'Ъ': 10}
word = input("Введите слово: ")
word_upper = word.upper()
total_score = 0
if word_upper[0] in english_scores:
        scores = english_scores
else:
        scores = russian_scores
for letter in word_upper:
        total_score += scores[letter]
print(f"Стоимость слова '{word}' = {total_score} очков")
```

```python
cats = [("Муся", 7, "Игорь", "Бероев"),
    ("Изольда", 2, "Игорь", "Бероев"),
    ("Барсик", 5, "Ольга", "Смирнова"),
    ("Мурка", 3, "Игорь", "Бероев"),
    ("Рыжик", 4, "Анна", "Петрова"),
    ("Снежок", 6, "Ольга", "Смирнова"),]
owners_cats = {}
for name, age, owner_name, owner_surname in cats:
    key = (owner_name, owner_surname)
    cat_info = f"{name}, {age}"
    if key not in owners_cats:
        owners_cats[key] = []
    owners_cats[key].append(cat_info)
for (owner_name, owner_surname), cat_info in owners_cats.items():
    print(f"{owner_name} {owner_surname}: {', '.join(cat_info)}")
```
