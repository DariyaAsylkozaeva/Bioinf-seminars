```python
import random
secret=random.randint(1,100)
attempts=10
print('Я загадал число от 1 до 100. У тебя 10 попыток: ')
for attempt in range(1, attempts+1):
    guess=int(input("Попытка {attempt}: "))
    if guess==secret:
        print("Ты выиграл")
        break
    elif guess<secret:
        print("Число больше")  
    else:
        print("Число меньше")    
print("Ты проиграл")      
```

```python
expr=input("Введите выражение: ").strip()
parts=expr.split()
a=float(parts[0])
b=float(parts[2])
op=(parts[1])
if op=="+":
    print(a+b)
elif op=="-":
    print(a-b)
elif op=="*":
    print(a*b)
elif op=="/":
    print(a/b)
```

```python
start=int(input("от: "))
end=int(input("до: "))
for num in range(start,end+1):
    if num<2:
        continue
    a=True
    for d in range(2,num):
        if num%d==0:
            a=False
            break
        
    if a:
        print(num)
    
```
